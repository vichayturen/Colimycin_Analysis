### 任务1
- 入院时间或出院时间包含在2020-12-01至2023-12-31期间
- 科室诊断或ICU诊断包含以下字样之一：脓毒症、脓毒血症、败血症等字样的行
- 对名字进行去重，只保留第一行
- 年龄大于等于18岁


```python
df = df3
df = df[
    (pd.to_datetime(df['入病区日期']).between('2020-12-01', '2023-12-31') | pd.to_datetime(df['出病区日期'], errors='coerce').between('2020-12-01', '2023-12-31'))
]
print(df.shape)
df = df[
    df['科室诊断'].str.contains('脓毒症|脓毒血症|败血症') | df['ICU诊断'].str.contains('脓毒症|脓毒血症|败血症')
]
print(df.shape)
df['年龄'] = df['年龄'].str.replace('岁', '').astype(int)
df = df[
    df['年龄'] >= 18
]
print(df.shape)
```
